## Objective

When modeling # of 311 issues, a big confounding factor is these non-residential issues, esp when `y` is # of issues per capita. There are 2 possible solutions to this:

1. Make separate models for non-residential TYPEs and residential TYPEs.
2. Make separate models for non-residential areas and residential areas.

I'd prefer to answer #1, as that way I wouldn't exclude whole areas where some people live (like downtown). Between residential and non-residential, I care more about the latter, and I almost don't care at all about the former, since I want to see what's associated with socioeconomic indicators.
