## Objective

The idea is that different groups of TYPEs interact differently with the features. Using my 'domain knowledge', I'll group the TYPEs together based on how I think they'll interact with the features differently.

```python
aa = df[df.num_issues < df.num_issues.quantile(0.70)].drop('TYPE', axis=1).groupby('REASON').sum().sort_values('num_issues', ascending=False).reset_index()
bb = aa[aa.num_issues > df.num_issues.quantile(0.70)].REASON
print bb.shape
bb.head()
```
