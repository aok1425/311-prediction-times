## What % of my best-performing categs have descs?

```python
categs_best_10 = set(['Unsanitary Conditions - Employees',
 'Schedule a Bulk Item Pickup SS',
 'Food Alert - Confirmed',
 'Notification',
 'Schedule a Bulk Item Pickup',
 'Unsanitary Conditions - Food',
 'OCR Front Desk Interactions',
 'Unsanitary Conditions - Establishment',
 'Abandoned Vehicles',
 'Recycling Cart Inquiry'])
```

```python
def count_desc_null_pct(df_mini):
    num_nulls = df_mini.description.isnull().sum()
    num = df_mini.shape[0]
    new_df = df_mini.count().T
#     new_df.columns = ['total', 'num_w_descs']
#     import pdb; pdb.set_trace()
#     new_df['pct'] = new_df.num_w_descs / new_df.total
    return new_df.T
    return num_nulls / num
```

```python
new_df = df_orig[['TYPE', 'description']].groupby('TYPE').apply(count_desc_null_pct)
new_df.columns = ['total', 'num_w_descs']
new_df['pct'] = new_df.num_w_descs / new_df.total
new_df = new_df.sort_values('pct').reset_index()
new_df
```
