## Zipcode

What are the number of issues by zip? Over time?

```python
df[['LOCATION_ZIPCODE', 'CASE_ENQUIRY_ID']].head(100).groupby(['LOCATION_ZIPCODE']).count() \
    .reset_index().sort_values(['LOCATION_ZIPCODE', 'CASE_ENQUIRY_ID'], ascending=False)
```

## Zipcode and Reason

```python
df[['LOCATION_ZIPCODE', 'REASON', 'CASE_ENQUIRY_ID']].head(100).groupby(['LOCATION_ZIPCODE', 'REASON']).count() \
    .reset_index().sort_values(['LOCATION_ZIPCODE', 'CASE_ENQUIRY_ID'], ascending=False)
```
